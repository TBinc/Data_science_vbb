```python
day_week = {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}

import math

def is_paired(event_id):
    if isinstance(event_id, str):
        return "Y"
    else:
        return "N"
```

```python
column_names = ["Program", "Paired", "Day", "Time_EST"]
data = pd.DataFrame(columns = column_names)

data_sessions = data_sessions[pd.to_numeric(data_sessions['MSM'], errors='coerce').notnull()]
data_sessions = data_sessions[~pd.to_numeric(data_sessions['COMPUTER_EMAIL'], errors='coerce').notnull()]
data['Program'] = data_sessions['COMPUTER_EMAIL'].astype(str).str[0:4]
# Drop 'http' from program
data = data[data['Program'].str.contains("http") == False]

data["Paired"] = data_sessions['EVENT_ID'].apply(lambda x: is_paired(x))

tmp = data_sessions['MSM'].astype(int)/1440
data['Time_EST'] = (data_sessions['MSM'].astype(int) - 1440*tmp.round())/60
data['Time_EST'] = data['Time_EST'].mod(24).astype(int)

data['Day'] = tmp.round().astype(int).mod(7).replace(day_week)
```

```python
data.to_csv('slots.csv', index=False)
```
